#Lab Objective:
1. To simulate a population distribution i.e uniform or exponential
2. To take multiple random samples of a given size from this population and compute their means
3. To observe how the distribution of these samoles means approches a normal distribution ass the number of samples and sample size increase


#Required Libraries:
1. Numpy: For data manipulation and taking random samples
2. Matplotlib: For visualization
3. Seaborn: For enhanced visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set random seed for reproductibility
np.random.seed(123)

# Set population distribution
population_size= 100000
#sample sizes to demonstrate convergence
sample_sizes= [1,5,10,40,100,200,400]
num_samples = 1000 #no of samples for each sample size

population = np.random.exponential(scale=3, size=population_size)

#Exponential Distribution
plt.figure(figsize=(10,6))
sns.histplot(population, bins=50, kde=True , color= 'blue', alpha=0.4)
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Demonstration of ClT
plt.figure(figsize=(14,10))

#Plot the distribution of sample means
for i, size in enumerate(sample_sizes, start=1):
  sample_means = [np.mean(np.random.choice(population,size=size, replace=False))
  for _ in range(num_samples)]

  plt.subplot(3,3,i)
  sns.histplot(sample_means, bins=30, kde=True , color='Red', alpha=0.4)
  plt.title(f'Sample mean Distribution(n={size})')
  plt.xlabel('Sample Mean')
  plt.ylabel('Frequency')

#Show the final plot
plt.suptitle('Demonstration of the Central Limit Theorem', fontsize=18)
plt.tight_layout()
plt.show()

#Observation
print("As sample size increase, the  sample means distribution converges to a normal distribution, regradless of the population's original distribution.")
```
