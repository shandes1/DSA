#Hypothesis testing
-> In this lab, we test hypothesis using z test, t test and F test With one or more means

```python
import numpy as np
import pandas as pd
from scipy import stats
```

#Z test with one mean
A factory claims that the average weight of a certain type of packaged flour is 2 kg. A random sample of 30 packages is taken, and the average weight of the sample is found to be 1.98 kg with a standard deviation of 0.1 kg. At a significance level of 0.05, is there enough evidence to reject the factory's claim?

```python
pop_mean = 2 #factory clamed mean
pop_sd =0.1

np.random.seed(23)
packages=np.random.normal(loc=1.98 ,scale =0.1,size=30)

#caluclating sample mean and size
sample_mean=np.mean(packages)
sample_size=len(packages)

#z-score
z_score_calc= (sample_mean- pop_mean)/(pop_sd/np.sqrt(sample_size))

#critical z-value for left-tailed test
alpha=0.05
z_critical_left=stats.norm.ppf(alpha)

#result
#print(sample_mean)
print(f'Calculated z-score: {z_score_calc:.3f}')
print(f'Critical z-value: {z_critical_left:.3f}')

#compare
if z_score_calc< z_critical_left:
  print('The null hypothesis is rejected. The weight is less then 2 kg as the factory clamed')
else:
  print('The null hypothesis is acceped.The weight is about 2kg as the factory clamed.')
```

#z-test with two mean
A researcher wants to determine if there is a significant difference in the average scores of two groups of students after using different teaching methods.


*   Group 1 consists of 40 students and has a mean score of 75 with a standard deviation of 10.
*   Group 2 consists of 35 students and has a mean score of 80 with a standard deviation of 12.




At a significance level of α=0.05, is there evidence to suggest that the teaching methods produce different average scores?

```python
#np.random.seed(23)
group1= np.random.normal(loc=75, scale=10, size=40)
group2= np.random.normal(loc=80, scale=80 , size=35)

#group 1
group1_mean= np.mean(group1)
group1_std =np.std(group1, ddof=1)
group1_size=len(group1)

#group 2
group2_mean= np.mean(group2)
group2_std =np.std(group2, ddof=1)
group2_size=len(group2)

#z-score
z_score_calc= (group1_mean - group2_mean)/np.sqrt(((group1_std**2)/group1_size)+((group2_std**2)/group2_size))

#critical z-score for two tailed
alpha=0.05
z_critical = stats.norm.ppf(1-alpha/2)

#result
print( group1_mean,group2_mean)
print(f'z-score: {z_score_calc:.3f}')
print(f'Critical z-value: {z_critical:.3f}')

#compare
if z_score_calc> z_critical or z_score_calc< -z_critical:
  print('Reject the null hypothesis.There is a significiant difference in the avarage score of two groups.')
else:
  print('Accept the null hypothesis. There is no significant difference in the average score of two groups.   ')
```

#t-test with single mean
A researcher claims that the average height of adult women in a city is 160 cm. To verify this claim, a sample of 25 women is taken, and their average height is found to be 162 cm with a sample standard deviation of 8 cm. At a significance level of
α=0.05, does this sample provide enough evidence to reject the claim that the average height is 160 cm?

```python
np.random.seed(123)
women=np.random.normal(loc=162,scale=8,size=25)

pop_mean=160
sample_mean= np.mean(women)
sample_std= np.std(women, ddof=1)
sample_size=len(women)
df=sample_size -1

#t value calculated
t_calc, _=stats.ttest_1samp(women,popmean=pop_mean)

#critical t value
alpha=0.05
t_critical= stats.t.ppf(1- alpha/2,df)

#results
print(sample_mean)
print(f'T value: {t_calc}')
print(f'Critical t value: {t_critical}')

#compare
if abs(t_calc)> t_critical:
  print('Reject the null hypothesis. The avg height is found to be greater than 160 cm')
else:
  print('Accept the null hypothesis. The avg height is found to be around 160 cm')
```

#T-test with two mean
A medical researcher is testing the effectiveness of a new treatment. Fifty participants were randomly selected and divided into two groups: those who received the treatment (Yes) and those who did not (No). After the study, the success rates of participants were recorded, where success rates are numerical scores ranging from 0 to 100.

```python
np.random.seed(41)

data = {
    'Treatment': [np.random.choice(['Yes', 'No']) for _ in range(50)],
    'Success_rate': [np.random.randint(50, 100) for _ in range(50)]
}
#data frame
df= pd.DataFrame(data)
# print(df)

#seperate the data
yes= df[df['Treatment']=='Yes']['Success_rate']
no= df[df['Treatment']=='No']['Success_rate']

#mean and std
yes_mean=np.mean(yes)
yes_std = np.std(yes, ddof=1)
yes_size=len(yes)


no_mean=np.mean(no)
no_std= np.std(no, ddof=1)
no_size= len(no)

#calculated t value
variance= ((yes_std**2)/yes_size) + ((no_std**2)/no_size)
t_stat= (yes_mean - no_mean)/ np.sqrt(variance)

#critical t value
df= ((yes_std **2 /yes_size + no_std**2/no_size)**2) / \
  ((yes_std **2 /yes_size)**2/ (yes_size -1)+ (no_std**2/no_size)**2 / (no_size -1))

alpha = 0.05
t_critical = stats.t.ppf(1- alpha/2, df)

#results
print(f'T_value: {t_stat}')
print(f'T_critical: {t_critical}')
print(f"Degrees of Freedom: {df}")

#compare
if abs(t_stat)> t_critical:
  print("Reject the null hypothesis. There is a significant difference in success rates between the two groups.")
else:
    print("Accept the null hypothesis. There is no significant difference in success rates between the two groups.")
```

#F test
A company wants to analyze the impact of different marketing strategies on the amount spent by customers. The marketing strategies tested are Social Media Ads, Email Campaigns, and Influencer Partnerships. The company collects data on the amount spent by 30 customers in each group.Significance level:
α=0.05

```python
#data
data={
    "Marketing Strategy":[np.random.choice(['Social media ads','Email campaigns',
                                            'Influencer Partnerships']) for _ in range(150)],
    'Amount Spend':[np.random.randint(100,2000) for _ in range(150)]
}

#data frame
df=pd.DataFrame(data)

#seperate the data
social_media_ads = df[df['Marketing Strategy']== 'Social media ads']['Amount Spend']
Email_campaigns = df[df['Marketing Strategy']== 'Email campaigns']['Amount Spend']
Influencer_Partnerships = df[df['Marketing Strategy']== 'Influencer Partnerships']['Amount Spend']

t=3
n=len(df)

#degree of freedom
df_between= t-1
df_within= n-t

#ANOVA test
f_stat,_ = stats.f_oneway(social_media_ads,Email_campaigns,Influencer_Partnerships)

#critical f value
alpha=0.05
f_critical = stats.f.ppf(1-alpha, df_between , df_within)

#result
print(f'F-value: {f_stat}')
print(f'F critical: {f_critical}')

#compare
if f_stat> f_critical:
  print("Reject the null hypothesis: There are significant differences between the group means.")
else:
    print("Accept the null hypothesis: There are no significant differences between the group means.")
```
